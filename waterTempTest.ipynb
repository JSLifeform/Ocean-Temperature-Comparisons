```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Feb 20 16:38:45 2021

@author: pchavan
"""


import re
import pandas as pd
import os


#snippet to split files and dump in a folder labeled data/
def files():
    n = 0
    while True:
        n += 1
        yield open('data/%d.csv' % n, 'w')


pat = '#--------------------------------------------------------------------------------'

fs = files()
outfile = next(fs) 
filenam='data/waterTemp2008.csv'
with open(filenam) as infile:
    for line in infile:
        if pat not in line:
            outfile.write(line)
        else:
            items = line.split(pat)
            outfile.write(items[0])
            for item in items[1:]:
                outfile = next(fs)
                outfile.write(pat + item)
                
outfile.close()
#snippet to further take every file in the data/ folder and then only read lines latitude, longitude and the table as that contains temperature       
START_PATTERN = 'UNITS'
END_PATTERN = 'END OF VARIABLES SECTION'
lat='Latitude'
long='Longitude'
for filename in os.listdir('data/'):
    if filename != ".DS_Store":
        with open('data/'+filename, mode="r", encoding="utf-8") as file:
            print(filename)
            match = False
            newfile = None

            for line in file:
                if re.match(lat, line):
                    match = True
                    newfile = open('tabular_data/'+filename, 'w')
                    newfile.write(line)
                    
                
                elif re.match(long, line):
                    match = False
                    newfile.write(line)
                
                elif re.match(START_PATTERN, line):
                    match = True
                    continue
                
                elif re.match(END_PATTERN, line):
                    match = False
                    newfile.close()
                    continue
                elif match:
                    newfile.write(line)
                    newfile.write('\n')
      
```
